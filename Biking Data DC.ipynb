```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bikeDays = pd.read_csv("http://knuth.luther.edu/~leekent/biking/hour.csv")
```

# Box and Whisker Plot

```python
data = [bikeDays["temp"], bikeDays["atemp"], bikeDays["hum"], bikeDays["windspeed"]]
```

```python
plt.boxplot(data, False, "rs", False)
plt.show()
```

# Bubble Plot

Let's find the peak time during workdays for bike riders

```python
workdayAveragesByHour = bikeDays[bikeDays.workingday==1].groupby('hr').mean()["cnt"]
workdayAveragesByHour
```

```python
workdayAveragesByHour.plot(kind="bar", width=0.8, figsize=(10,5), color = 'blue')
plt.show()
```

```python
workdayAveragesBySeasonHour = bikeDays[bikeDays.workingday==1].groupby(["season","hr"]).mean()
workdayAveragesBySeasonHour 
```

```python
workdayAveragesBySeasonHour.reset_index(level=0,inplace=True)
workdayAveragesBySeasonHour
```

```python
workdayAveragesBySeasonHour.reset_index(level=0,inplace=True)
workdayAveragesBySeasonHour
```

```python
workDaysByMonthHour = bikeDays[bikeDays.workingday==1].groupby(['mnth','hr']).mean()
workDaysByMonthHour.reset_index(level=0,inplace=True)
workDaysByMonthHour.reset_index(level=0,inplace=True)
workDaysByMonthHour
```

```python
workDaysByMonthHour.plot.scatter(x='mnth', y='hr', s=workDaysByMonthHour['cnt'], color = "lightgreen");
```

```python
workdayAveragesBySeasonHour.plot.scatter(x="season",y="hr", s=workdayAveragesBySeasonHour["cnt"], color=workdayAveragesBySeasonHour["season"])
plt.show()
```

# Violin Plot

```python
heatMapData = workDaysByMonthHour[["mnth","hr","cnt"]].pivot("hr","mnth","cnt")
heatMapData
```
