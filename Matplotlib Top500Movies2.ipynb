# Finding the Adjusted Gross by Year per Production Company (the top 4 only)

To do this efficiently we will join two dataframs and then we will need to gorup by and sum to get the total. In the end I's like to display as stacked area chart

```python
df = pd.DataFrame(np.random.rand(10,4), columns =["a", "b", "c", "d"])
```

```python
df
```

```python
df.plot.area()
```

```python
top4 = movieProd.groupby(["Production"]).sum().sort_values("AdjustedGross", ascending = False).head(4)
top4
```

```python
top4.index
```

```python
boMovies.index
```

```python
top4ProdMovies = boMovies.loc[boMovies["Production"].isin(top4.index)]
```

```python
top4ProdMovies
```

```python
top4YearAdjustedGross = top4ProdMovies[["Production", "Year", "AdjustedGross"]]
top4YearAdjustedGross
```

```python
top4YearAdjustedGross.set_index("Production", inplace=True)
```

```python
top4YearAdjustedGross
```

```python
top4ByYear = top4YearAdjustedGross.groupby(["Production", "Year"]).sum()
top4ByYear
```

```python
top4ByYear.reset_index(level = 0, inplace = True)
top4ByYear
```

```python
top4ByYearP = top4ByYear.pivot_table(values="AdjustedGross", index = top4ByYear.index, columns = "Production")
top4ByYearP
```

```python
top4ByYearP.fillna(0,inplace=True)
```

```python
top4ByYearP
```

```python
top4ByYearP.plot.area()
plt.show()
```

```python
top4ByYearP.index = top4ByYearP.index.astype(int)
```

```python
top4ByYearSince1985 = top4ByYearP[top4ByYearP.index > 1985]
top4ByYearSince1985
```

```python
top4ByYearSince1985.plot.area()
plt.show()
```
