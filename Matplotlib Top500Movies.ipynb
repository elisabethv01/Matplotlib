```python
#Extract BoxOffice values that is > 0
categorical = movies.BoxOffice > 0
BOMovies = movies[categorical]

#Extract adjusted_boxOffice values that is > 0
criterial = movies.adjusted_boxOffice > 0
boMovies = movies[criterial]

plt.plot("Year","adjusted_boxOffice",'ro', color="black", data=boMovies)
plt.plot("Year","BoxOffice", 'ro',color="black", data = BOMovies, alpha = 0.4)
plt.title('Box Office')
plt.xlabel('Year')
plt.ylabel('BoxOffice (00s Million USD)')
plt.legend()
plt.show()
```

```python
byYear = movies["Year"].value_counts()
plt.bar(byYear.index, byYear, color='brown')
plt.title('Movie Production')
plt.xlabel('Year')
plt.ylabel('Movie Produced')
plt.show()
```

```python
#Extract adjusted_boxOffice values that is > 0
criterial = movies.adjusted_boxOffice > 0
boMovies = movies[criterial]

fig, ax = plt.subplots(figsize=(10,5)) #creates two objects
ax.plot(boMovies['Year'], boMovies['adjusted_boxOffice'],'ro', color = 'black')
ax.plot(boMovies['Year'], boMovies['BoxOffice'],'ro', color = 'black', alpha = 0.4)
ax.set_xlabel('Year')
ax.set_ylabel('BoxOffice (00s Million USD)')
ax.set_title('Scatter Plot')
ax.legend()
fig.suptitle('BoxOffice and Adjusted BoxOffice', size=15, y=1.03)
```
